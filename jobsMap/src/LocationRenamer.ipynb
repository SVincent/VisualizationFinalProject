```python
import re

# replaces the location description with the state (if identifyable)
def repl_location(loc):
    for key in states:
        if (key in loc.lower() or states[key] in loc):
            return key
    return "FALSE"

# replaces the jobtype with one of five categories
def repl_jobtype(jobtype):
    if ("employee" in jobtype.lower()):
        if ("full time" in jobtype.lower()):
            return 'full time employee'
        elif ("part time" in jobtype.lower()):
            return 'part time employee'
        else:
            return 'other'
    elif ("contract" in jobtype.lower()):
        if ("full time" in jobtype.lower()):
            return 'full time contractor'
        else:
            return 'other'
    elif (jobtype == None):
        return 'None'
    else:
        return 'other'

# parses the salary
def parse_salary(salary):
    yearly = False
    lower = None
    upper = None
    
    if ("year" in salary.lower()):
        yearly = True
        lower = re.search("^[0-9]*[,]*[0-9]+[.]00.[-]", salary)
        if (lower != None):
            lower = lower.group()[:-2]
            lower = float(lower.replace(',', ''))
        upper = re.search("[0-9]*[,]*[0-9]+[.]00.[$]", salary)
        if (upper != None):
            upper = upper.group()[:-2]
            upper = float(upper.replace(',', ''))
    elif ("hour" in salary.lower()):
        lower = re.search("^[0-9]+[.]00.[-]", salary)
        if (lower != None):
            lower = lower.group()[:-2]
            lower = float(lower.replace(',', ''))
        upper = re.search("[0-9]+[.]00.[$]", salary)
        if (upper != None):
            upper = upper.group()[:-2]
            upper = float(upper.replace(',', ''))
    return [yearly, lower, upper]
```

```python
import csv

# The dirty/uncleaned dataset
infile = '/home/vincent/Documents/programming/VisualizationFinalProject/jobsMap/src/dataset/us-jobs-on-monstercom/monster_com-job_sample.csv'

# open the CSV and run over all rows, calling the various replacement and parsing functions
f = open(infile)
csv_f = csv.reader(f)

rowcount = 0
rows = []

for row in csv_f:
    newrow = []
    if (rowcount != 0):
        #replace location
        row[8] = repl_location(row[8])
        if (row[8] != "FALSE"):
            newrow.append(row[8])
            newrow.append(statecodes[row[8]])
        #replace jobtype
        newrow.append(repl_jobtype(row[7]))
        #parse salary
        parsed_salary = parse_salary(row[11])
        newrow.append(parsed_salary[0])
        newrow.append(parsed_salary[1])
        newrow.append(parsed_salary[2])
        
        #add organization, sector and job title to new row
        newrow.append(row[9].replace(',', ''))
        newrow.append(row[12].replace(',', ''))
        newrow.append(row[6].replace(',', ''))
        
        #add newrow to rows to be written into the cleaned dataset
        rows.append(newrow)
    rowcount += 1
```

```python
outfile = '/home/vincent/Documents/programming/VisualizationFinalProject/jobsMap/src/dataset/us-jobs-on-monstercom/cleaned_data.csv'

fields = ['location', 'id', 'job_type', 'yearly', 'salary_lower', 'salary_upper', 'organization', 'sector', 'job_title']

# writing to csv file 
with open(outfile, 'w') as csvfile: 
    # creating a csv writer object 
    csvwriter = csv.writer(csvfile) 
      
    # writing the fields 
    csvwriter.writerow(fields) 
      
    # writing the data rows 
    csvwriter.writerows(rows)
```
