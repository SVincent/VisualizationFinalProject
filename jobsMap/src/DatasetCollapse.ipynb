```python
import csv
import re
from collections import Counter

infile = '/home/vincent/Documents/programming/VisualizationFinalProject/jobsMap/src/dataset/us-jobs-on-monstercom/cleaned_data.csv'

f = open(infile)
csv_f = csv.reader(f)

rowcount = 0
rows = []

total_jobs = [0] * 50
other_job_types = [0] * 50
fulltime_employees = [0] * 50
fulltime_contractors = [0] * 50
parttime_employees = [0] * 50
avg_yearly_salary = [0] * 50
avg_hourly_salary = [0] * 50
salary_lower_sum = [0] * 50
salary_lower_count = [0] * 50
salary_upper_sum = [0] * 50
salary_upper_count = [0] * 50
wage_lower_sum = [0] * 50
wage_lower_count = [0] * 50
wage_upper_sum = [0] * 50
wage_upper_count = [0] * 50
#sectors as list of lists
sectors = [[] for x in range(0,50)]

for row in csv_f:
    if (rowcount != 0):
        state_id = int(row[14])-1
        if ("employee" in row[7].lower()):
            if ("full time" in row[7].lower()):
                fulltime_employees[state_id] += 1
                total_jobs[state_id] += 1
            elif ("part time" in row[7].lower()):
                parttime_employees[state_id] += 1
                total_jobs[state_id] += 1
            else:
                other_job_types[state_id] += 1
                total_jobs[state_id] += 1
        elif ("contract" in row[7].lower()):
            if ("full time" in row[7].lower()):
                fulltime_contractors[state_id] += 1
                total_jobs[state_id] += 1
            else:
                other_job_types[state_id] += 1
                total_jobs[state_id] += 1
                
        if ("year" in row[11].lower()):
            lower = re.search("^[0-9]*[,]*[0-9]+[.]00.[-]", row[11])
            if (lower != None):
                lower = lower.group()[:-5]
                lower = int(lower.replace(',', ''))
                salary_lower_sum[state_id] += lower
                salary_lower_count[state_id] += 1
                #print(lower)
            upper = re.search("[0-9]*[,]*[0-9]+[.]00.[$]", row[11])
            if (upper != None):
                upper = upper.group()[:-5]
                upper = int(upper.replace(',', ''))
                salary_upper_sum[state_id] += upper
                salary_upper_count[state_id] += 1
                #print(upper)
        elif ("hour" in row[11].lower()):
            lower = re.search("^[0-9]+[.]00.[-]", row[11])
            if (lower != None):
                lower = lower.group()[:-2]
                lower = float(lower.replace(',', ''))
                wage_lower_sum[state_id] += lower
                wage_lower_count[state_id] += 1
                #print(lower)
            upper = re.search("[0-9]+[.]00.[$]", row[11])
            if (upper != None):
                upper = upper.group()[:-2]
                upper = float(upper.replace(',', ''))
                wage_upper_sum[state_id] += upper
                wage_upper_count[state_id] += 1
                #print(upper)
        
        if (row[12] != None):
            sectors[state_id].append(row[12].lower())
    rowcount += 1
    
sector = []
for i in range(0, 50):
    sector.append(Counter(sectors[i]))
print(sector)
    
for i in range(0, 50):
    row = []
    row.append(i+1)
    row.append(inv_map[i+1])
    row.append(total_jobs[i])
    row.append(other_job_types[i])
    row.append(fulltime_employees[i])
    row.append(fulltime_contractors[i])
    row.append(parttime_employees[i])
    avg_lower_salary = salary_lower_sum[i]/max(salary_lower_count[i],1)
    avg_upper_salary = salary_upper_sum[i]/max(salary_upper_count[i],1)
    row.append(avg_lower_salary)
    row.append(avg_upper_salary)
    row.append((avg_lower_salary+avg_upper_salary)/2)
    avg_lower_wage = wage_lower_sum[i]/max(wage_lower_count[i],1)
    avg_upper_wage = wage_upper_sum[i]/max(wage_upper_count[i],1)
    row.append(avg_lower_wage)
    row.append(avg_upper_wage)
    row.append((avg_lower_wage+avg_upper_wage)/2)
    rows.append(row)
```
